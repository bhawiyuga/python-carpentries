# Visualizing Tabular Data


### Objectives
- Plot simple graphs from data

## Visualizing patient data

- The best way to develop insight is often to visualize data
- `matplotlib` is the de facto standard plotting library in Python
- We'll use the `pyplot` module from `matplotlib` to create visualizations (e.g. heatmaps, line)

**Heat maps:**
- Use `matplotlib.pyplot.imshow()` to create a heat map of the data
- Each row represents a patient, each column represents a day
- Blue pixels represent low values, yellow pixels represent high values
- Helps identify patterns in the data quickly

```python
import matplotlib
```

```python
# Import pyplot and use the alias plt
import matplotlib.pyplot as plt
```

```python
import numpy as np

data = np.loadtxt(fname="inflammation-01.csv", delimiter=",")
```

```python
heatmap_image = plt.imshow(data)
plt.show()
```

**Line plots:**
- Use `matplotlib.pyplot.plot()` to create line graphs
- Can visualize statistics over time (average, maximum, minimum)
- `numpy.mean(data, axis=0)` - calculates average inflammation per day
- `numpy.amax(data, axis=0)` - finds maximum inflammation per day
- `numpy.amin(data, axis=0)` - finds minimum inflammation per day

```python
# Calculate average inflammation for all patients per-day
average_per_day = np.mean(data, axis=0)
print(average_per_day.shape)
```

```python
# Make line plot of average inflammation per day
ave_plot = plt.plot(average_per_day)
plt.show()
```

```python
max_per_day = np.amax(data, axis=0)
plt.plot(max_per_day)
plt.show()
```

```python
min_per_day = np.amin(data, axis=0)
plt.plot(min_per_day)
plt.show()
```

**Key observations:**
- Look at average, maximum, and minimum values, not just one statistic
- Visualization can reveal unexpected patterns that would be difficult to see in raw numbers
- Suspicious patterns (e.g., perfectly linear trends) may indicate data problems

## Grouping plots

**Organizing multiple plots:**
- Use `matplotlib.pyplot.figure()` to create a figure space
  - `figsize` parameter controls the size (e.g., `figsize=(10.0, 3.0)`)
- Use `fig.add_subplot()` to add individual plots
  - Takes 3 parameters: (rows, columns, plot_number)
  - Example: `add_subplot(1, 3, 1)` means 1 row, 3 columns, first plot

```python
# Define figure and axes with size 10 inches wide by 3 inches tall
fig = plt.figure(figsize= (10,3))

# Define axes for three sub-plots in one row
ax1 = fig.add_subplot(1, 3, 1) # num of rows, num of columns, plot number
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

# Define the labels for y-axis
ax1.set_ylabel('average')
# Plot the average inflammation per day on ax1
ax1.plot(average_per_day)

ax2.set_ylabel('max')
ax2.plot(max_per_day)

ax3.set_ylabel('min')
ax3.plot(min_per_day)

# Adjust the padding between and around sub-plots
fig.tight_layout()
plt.savefig('inflammation-01-summary.png')
plt.show()
```

**Customizing subplots:**
- Use `set_xlabel()` and `set_ylabel()` to label axes
- Use `set_ylim(min, max)` to set y-axis range
- Use `fig.tight_layout()` to prevent overlapping plots

**Saving figures:**
- Use `matplotlib.pyplot.savefig('filename.png')` to save plots
- Matplotlib automatically determines format from file extension (PNG, SVG, PDF, JPEG)

**Import shortcuts:**
- Common practice: `import matplotlib.pyplot as plt` and `import numpy as np`
- Shorter code but requires consistency throughout your program
