# Analyzing Data from Multiple Files

Objectives

- Use a library function to get a list of filenames that match a wildcard pattern.
- Write a for loop to process multiple files.

Suppose we want to process multiple files. Then, we need to list multiple files within a directory. To do that, we can use `glob` module.

```python
import glob
```

```python
inflammation_files = glob.glob("inflammation-*.csv")
print(inflammation_files)
```

## Load and visualize multiple data files

We would like to load all inflammation input files. Then, we will visualize the max, min, and average for all days.

```python
sorted_files = sorted(inflammation_files)
sorted_files
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Lets visualize min, max, and average for all files
for filename in sorted_files:
    print(filename)
    # Load the data
    data = np.loadtxt(fname=filename, delimiter=',')
    # Define figures
    fig = plt.figure(figsize=(10.0, 3.0))
    # Define axes for min, max, average
    ax1 = fig.add_subplot(1, 3, 1)
    ax2 = fig.add_subplot(1, 3, 2)
    ax3 = fig.add_subplot(1, 3, 3)
    # Define y-axis labels
    ax1.set_ylabel('average')
    ax2.set_ylabel('max')
    ax3.set_ylabel('min')
    # Plot data min, max, average on column-wise (per-day basis)
    ax1.plot(np.mean(data, axis=0))
    ax2.plot(np.amax(data, axis=0))
    ax3.plot(np.amin(data, axis=0))
    # Put tight layout
    plt.tight_layout()
    # Show the plot
    plt.show()
```
